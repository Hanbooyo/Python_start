```python
import matplotlib.pyplot as plt  # matplotlib의 pyplot을 앞으로 plt라고 하겠다는 선언

X = ['M', 'T', 'W', 'T', 'F', 'S', 'S']
Y1 = [15, 14, 16, 18, 17, 15, 19]
Y2 = [20, 23, 23.5 ,25, 23, 24, 26]


plt.plot(X,Y1, label="Seoul")
plt.plot(X,Y2, label="Busan")
plt.xlabel("day")
plt.ylabel("temperature")
plt.legend(loc="upper left") # 왼쪽 위에 띄워줌
plt.title("Temperatures of Cities")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

flights = sns.load_dataset("flights")

plt.figure(figsize=(12,3)) #행, 열 사이즈 설정

print(flights)
```

```python
sns.barplot(data = flights, x='year', y='passengers')
```

```python
sns.boxplot(data = flights, x='year', y='passengers') # 박스플롯을 이용한 시각화 ( 신뢰구간 )
```

```python
sns.lineplot(flights["passengers"]) # 라인을 이용한 (신뢰구간표시)
```
